```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt

# plt相關參數設定
plt.rcParams['font.sans-serif'] = ['simhei'] # 可以顯示中文字
plt.figure(figsize=(6,6)) # 畫布大小

# 要讀取的相片名稱(必須放在和該程式同一個目錄下)
# img_info = 'example_image_for_face_recognition_2.jpg'
img_info = 'example_image_for_face_recognition.jpg'

# 載入分類器
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
# 讀取圖片
img = cv2.imread(img_info)
# 印出原始圖片
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('原始圖片', fontsize=18)
plt.show()

# 轉成灰階圖片
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 偵測臉部
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.08,
    minNeighbors=5,
    minSize=(32, 32))
# 繪製人臉部份的方框
counter = 1 # 現在是第幾張人臉
for (x, y, w, h) in faces:
    # 裁切圖片
    crop_img = img[y:y+h, x:x+w]
    # 保存圖片
    cv2.imwrite(f"{img_info}_faceOnly_{counter}.jpg", crop_img)
    counter += 1
    # 打印圖片
    plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
    plt.title('臉部偵測結果', fontsize=18)
    plt.show()
    
```
